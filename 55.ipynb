```python
#pandaas
import pandas as pd
import numpy as np

## SERIES
# series in pandas is a one dimensional array type object which can hold any datatype. 

data = [1,2,3,4,5]
print(pd.Series(data))

#create a series from dictionary 
diction = {'a' : 1, 'b':2, 'c':3}
print("Series from a dictionary\n ",pd.Series(diction)) #returns a new pandas series.
print(diction)

#creating series with custom indexes
indexes = ['x','z','y','k','m']
print("Pandas series with custom data", pd.Series(data, index=indexes))

```

```python
# DataFrame
# Creating a DataFrame from a dictionary
# Dataframe is a 2D array object which is represented as rows and cols in a table. it contains indexes and headings of cols
# This is how data is there in an excel sheet

diction = {'city': ['Bengaluru', 'Mysuru', 'udupi'],
            'fruits' : ['Apple', 'Banana', 'Strawberry']}

print("Pandas dataframe from the dictionary of lists\n", pd.DataFrame(diction))

#Dataframe can be converted into a 2D numpy array
print("The numpy array of the dataframe is (ignoring column names and indexes\n", np.array(pd.DataFrame(diction)))

lists = [ {'city': 'mysuru', 'fruit' : 'apple', 'bike':'tvs'}, {'city': 'bengaluru', 'fruit':'banana', 'bike':'honda'}, {'city': 'dubai', 'fruit':'guava','bike': 'bajaj'}]
print("Creating a Dataframe from a list of dictionaries\n", pd.DataFrame(lists))
```

```python
#Reading csv files
df = pd.read_csv('amazon.csv')
print(df.head(5))
```

```python
#Accessing Data from DataFrame
data = {'Name':['Anston', 'Yashita','Deepthi'], 'Age':[19,20,20], 'City':['Udupi', "Ballari", "Mysuru"]}
df = pd.DataFrame(data)
print(df)

#selecting a single column

print(df['Age'])

#selecting a row
print("First row\n", df.loc[0])
```

```python
# Accessing specific elements
# df.at [index, col. name]

print(df.at[1,'Name'])

# df.iat[index, index] this treats the df like a matrix and accesses elements like that

print("Element at position 1,1 is :",df.iat[1,1])
```

```python
# Data manipulation 

# we can add columns into the dataframe without using any functions

df['Sex'] = ['Male', 'Female', 'Female']
df["Age"] = [19,20,20]
print("updated df with added column\n", df)

# Remmoving a column
df.drop(columns = "Sex", inplace=True)
print("Dataframe with the sex column dropped\n", df)

df['Age'] +=1

#dropping rows

df.drop(0, inplace=True)
df
```
