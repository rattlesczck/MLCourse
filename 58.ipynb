```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.plot(x,y, color = 'red', linestyle = '--', marker = 'o', linewidth = '3') #different parameters of the plot
plt.grid(True) #to add grids
```

```python
#creating multiple plots

plt.subplot(1,2,1) #parameters are (rows, columns, position)
#here we have 1 row, 2 cols.. and we're talking bout the graph in the first position
plt.plot(y,x, color = 'blue')
plt.title("First plot")

plt.subplot(1,2,2) #we're describing the 2nd graph
plt.plot(x,y, color = 'pink')
plt.title("Second plot")

# similarly we can have multiple plots with more than 1 rows/cols. we can have plt.subplot(2,2,1)

```

```python
#BAR GRAPHS

columns = ['A', 'B', 'C', 'D','E']
values = [1,6,4,2,12]

plt.bar(columns, values, color = 'red')
#rest of the parameters are the same as for normal graphs.. like xlabel, ylabel, title, color etc
```

```python
#Creating histograms

data = [1,2,2,3,4,5,5,6,7,8]

plt.hist(data, bins=5, color = 'pink', edgecolor = 'black') #creating a histogram with 5 bins.. i.e. 5 divisions
```

```python
#Scatter plot!!

x = [1,2,3,4,5,6]
y = [9,8,7,6,5,4]

plt.scatter(x,y,color = 'blue', marker = 'o') 
```

```python
# Pie chart

labels = ['A', 'B', 'C', 'D', 'E', 'F']
values = [30, 10, 12, 18, 10, 20]
explode = [0.2, 0,0,0,0.1,0]
colors = ['blue', 'pink', 'gold', 'red', 'yellow', 'green']

plt.pie(values, labels = labels, explode = explode, colors = colors, autopct = "%1.1f%%") #explode parameter is to define which part is to be seperated 
# from the pie chart



```

```python
#Example: Visualizing sales data
import pandas as pd

df = pd.read_csv('data.csv')
print(df.head())


```

```python
#plot total sales by products

df.info()

salesbyproduct = df.groupby('Category')['Sales'].sum()
print(salesbyproduct)
```

```python
salesbyproduct.plot(kind = 'bar', color = 'pink')
plt.xlabel("Product category")
plt.ylabel("Sales revenue")
```
