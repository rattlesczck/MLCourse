```python
#Seaborn

# its a python visualization library based on matplotlib

import seaborn as sns 
import matplotlib.pyplot as plt
#basic plotting with seaborn

tips = sns.load_dataset('tips')
print(tips)
```

```python
sns.scatterplot(x='total_bill', y = 'tip', data=tips) #parameters for the seaborn function
plt.title("Plot of total bill v/s tips")
plt.show()
```

```python
#Line plot with seaborn 


sns.lineplot(x='size', y = 'total_bill', data = tips)
plt.title("Line plot of size v/s total bill")
plt.show()
```

```python
#Categorical plots

#Bar plots

sns.barplot(x='day', y = 'total_bill', data = tips)
plt.title("Bar plot of total bill v/s day of the week")
plt.show()
```

```python
#Box plot

sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Box plot of days v/s total bill")
plt.show()
```

```python
#Violin plot (gives distribution of the data w.r.t. the total bill)
sns.violinplot(x='day', y = 'total_bill', data=tips)
plt.title("Violin plot of days v/s total bill")
plt.show()
```

```python
#Histograms

sns.histplot(tips['total_bill'], bins = 10, kde=True)
```
